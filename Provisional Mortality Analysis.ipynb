### 1. Is there a seasonal pattern in the number of deaths from these pathogens over time.?

```python
plt.figure(figsize=(13, 6))
sns.lineplot(x='week_end', y='percent_deaths', hue='pathogen', data=df)
plt.title('Trend Analysis of Pathogen Deaths Over Time')
plt.xlabel('Week End')
plt.ylabel('Percent Deaths')
plt.xticks(rotation=45)
plt.legend(title='Pathogen', loc='upper right')
plt.show()
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="lefAt"><font color=brown>üí¨ Inference:</font></h3>

1. <b>Consistent Influenza and RSV</b>: The analysis highlights the <b>consistency</b> in the percent deaths from Influenza and RSV over time. These pathogens maintain relatively <b>stable mortality rates</b> during the observed period.<br>
<br>
2. <b>Fluctuating COVID-19</b>: In contrast to Influenza and RSV, COVID-19 exhibits a <b>fluctuating trend</b> in percent deaths. The data shows <b>varying impacts</b> of COVID-19 on mortality rates during the same period.

----

### 2. What is the comparative impact of different pathogens on mortality rates?

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='pathogen', y='percent_deaths', data=df)
plt.title('Comparative Analysis of Pathogen Impact on Mortality Rates')
plt.xlabel('Pathogen')
plt.ylabel('Percent Deaths')
plt.show()
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="lefAt"><font color=brown>üí¨ Inference:</font></h3>

1. The <b>Impact of COVID-19</b>: The bar graph illustrates that the <b>impact of COVID-19</b> on mortality rates is <b>significant</b>, as the bar representing COVID-19 is the <b>largest</b> among the three pathogens. This suggests that COVID-19 has a substantial influence on mortality during the observed period.<br>
<br>

2. <b>Influenza's Influence</b>: In contrast, the bar representing Influenza is relatively <b>smaller</b>, indicating a <b>lesser impact</b> on mortality rates compared to COVID-19. It is worth noting that the bar for Influenza is visually smaller than COVID-19.<br>
<br>
3. <b>Negligible Impact of RSV</b>: The bar representing RSV is almost <b>negligible</b> in comparison to both COVID-19 and Influenza. This suggests that, during the observed period, RSV has a <b>minimal impact</b> on mortality rates, with the bar being barely visible on the graph.<br>

---

### 3. What are the mortality rate trends during transitional months?"

```python
df['month'] = df['week_end'].dt.month_name()
average_deaths_by_month = df.groupby('month')['percent_deaths'].mean().reset_index()

plt.figure(figsize=(14, 6))
plt.bar(average_deaths_by_month['month'], average_deaths_by_month['percent_deaths'], color='lightblue')
plt.title('Average Percent Deaths by Month')
plt.xlabel('Month')
plt.ylabel('Average Percent Deaths')
plt.xticks(rotation=45)
plt.show()
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="lefAt"><font color=brown>üí¨ Inference:</font></h3>

1. Seasonal Variation: Mortality rates exhibit a <b>seasonal pattern</b>, with higher percentages during <b>winter months</b> (January, December, February) and lower percentages in <b>summer months</b> (June, July, August).<br>
<br>
2. Low Mortality in Summer: Mortality rates are notably <b>lower during summer months</b>, indicating a potential relationship between warmer weather and reduced mortality.<br>
<br>
3. Transitional Periods: Months like March, April, May, September, October, and November show <b>intermediate mortality rates</b>, possibly representing transition periods between high winter mortality and low summer mortality.<br>
